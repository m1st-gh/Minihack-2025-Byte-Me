```python
from webbrowser import get
from tinydb import TinyDB, Query
import json
import pandas as pd

db = TinyDB('db.json')
source = db.table("water_sources")

with open('location_data.json', 'r') as file:
    source.insert_multiple(json.load(file))

def get_all_sensors():
    return source.all()

data = get_all_sensors()
display(pd.DataFrame((get_all_sensors())))
```
